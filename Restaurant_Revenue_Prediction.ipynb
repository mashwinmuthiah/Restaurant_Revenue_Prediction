### Reduce the dimension of the features using PCA

```python
#Standardizing the values of the features
from sklearn.preprocessing import StandardScaler
```

```python
features = train.columns[5:42]
features
```

```python
x = train.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(x).head()
```

```python
# Reducing the Dimentions of the features using PCA
#code below has .95 for the number of components parameter. 
#It means that scikit-learn choose the minimum number of principal components such that 95% of the variance is retained.

from sklearn.decomposition import PCA
pca = PCA(0.95)
```

```python
x = pd.DataFrame(pca.fit_transform(x))
```

```python
for i in np.arange(14):
    x.rename(columns={i:'PCA'+str(i+1)},inplace=True)
```

```python
a = train.iloc[:,0:5]
b = train.loc[:,'revenue']
final_df = pd.concat([a,x,b],axis=1)
```

```python
final_df.head()
```

```python
dates ={"Date":[],"Month":[],"Year":[]}
import datetime
for i in np.arange(137):
    date = datetime.datetime.strptime(final_df.Open_Date[i],"%m/%d/%Y")
    dates['Date'].append(date.day)
    dates['Month'].append(date.month)
    dates['Year'].append(date.year)
```

```python
final_df = pd.concat([final_df,pd.DataFrame(dates)],axis = 1)
```

```python
final_df.head()
```

```python
sns.boxplot(x='Month',y='revenue',data=final_df)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.boxplot(ax=ax1 ,x='City_Group',y='revenue',data = final_df)
sns.boxplot(ax = ax2 , x ='Type',y='revenue',data = final_df)
```

```python
city_list = final_df['City'].unique()
```

```python
dic = {'City_name':[],'Minimum_revenue':[],'Maximum_revenue':[],'Average_Revenue':[],"No_of_Rows":[]}
for i in np.arange(len(city_list)):
    a = pd.DataFrame(final_df[final_df.City==city_list[i]])
    dic["No_of_Rows"].append(len(a))
    dic['City_name'].append(city_list[i])
    dic['Minimum_revenue'].append(min(a.revenue))
    dic['Maximum_revenue'].append(max(a.revenue))
    dic['Average_Revenue'].append(np.mean(a.revenue))
```

```python
city_wise = pd.DataFrame(dic)
city_wise.head()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(city_wise.City_name,city_wise.Minimum_revenue)
ax2.plot(city_wise.City_name,city_wise.Maximum_revenue)
ax3.plot(city_wise.City_name,city_wise.Average_Revenue)
ax4.plot(city_wise.City_name,city_wise.No_of_Rows)
ax1.set_xticklabels(city_wise.City_name,rotation=90)
ax2.set_xticklabels(city_wise.City_name,rotation=90)
ax3.set_xticklabels(city_wise.City_name,rotation=90)
ax4.set_xticklabels(city_wise.City_name,rotation=90)
```
